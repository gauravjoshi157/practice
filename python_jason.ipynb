```python
# importing json module 
import json
```

```python
#opening json file 
# r is used for read file 
#encoding is used: utf-8
with open('sample.json','r',encoding = 'utf-8') as file:
    content = file.read()
    
```

```python
#load the data of the jason file in a variable named file_data
file_data = json.loads(content)
```

```python
#printing data
print(file_data)
```

```python
for x in file_data:
    print( f"Name: {x [ 'name']}")
    print( f"Code: {x [ 'code']}")
    print( f"Capital: {x [ 'capital']}")
    print( f"Region: {x [ 'region']}")
    print( f"Currency: {x ['currency']}")
    print( f"Language:{x['language']}")
    print( f"Flag:{x['flag']}")
    print( f"Dialling_code: {x['dialling_code']}")
    print( f"Iso_Code: {x['isoCode']}")
    print(f"Dialling_code: {x['dialling_code']}")
    
    
```
