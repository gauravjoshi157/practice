# Visualizing Global Temperature Changes

__Objective:__ Create a heatmap to visualize global temperature changes over the past century. Use a dataset containing average annual temperatures for different regions of the world.

__Data:__ You can use the Global Temperature Data for this task.

__Expectations:__

- Use a heatmap to show temperature changes over time.

- Add a color bar to indicate temperature ranges.

- Include a title and labels for axes.

__Hint:__ Use `sns.heatmap()` from the Seaborn library for creating the heatmap.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv(r'D:\internship\vs code\projects\datasets\global_temprature_data.csv')
data['dt'] = pd.to_datetime(data['dt'], errors='coerce')
# Pivot the data
data_pivot = data.pivot(index='dt', columns='Country', values='AverageTemperature')

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data_pivot, cmap='coolwarm', cbar_kws={'label': 'AverageTemperature (C)'})
plt.title('Global AverageTemperature Changes Over the Past Century')
plt.xlabel('Region')
plt.ylabel('Year')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv(r'D:\internship\vs code\projects\datasets\global_temprature_data.csv')
data['dt'] = pd.to_datetime(data['dt'], errors='coerce')

# Extract the year from the date column
data['year'] = data['dt'].dt.year

# Pivot the data, using the year for the index
data_pivot = data.pivot(index='year', columns='Country', values='AverageTemperature')

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data_pivot, cmap='coolwarm', cbar_kws={'label': 'Average Temperature (C)'})
plt.title('Global Average Temperature Changes Over the Past Century')
plt.xlabel('Year')
plt.ylabel('Region')

# Format the x-axis to show only the year
plt.xticks(rotation=90)
plt.show()
```
