Technology usage data including app usage, screen time, and demographic information. Download a suitable dataset from Kaggle, such as the Technology Adoption Dataset. OR use this data

**Visualization Goals**

- Analyze trends in technology usage over time.
- Compare usage patterns across different demographics.
- Visualize the impact of technology on productivity.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the dataset
# Replace with your actual file path
df = pd.read_csv(r"D:\internship\vs code\projects\datasets\technology.csv")

```

```python
# Inspect the column names to understand the structure
print("Columns in the dataset:", df.columns)

```

```python
# Convert 'year' to datetime format
df['year'] = pd.to_datetime(df['year'], format='%Y')

```

```python
# Set 'year' as the index
df.set_index('year', inplace=True)

```

```python
# Analyze trends in technology usage over time (assuming 'value' is screen time)
plt.figure(figsize=(10, 6))
df['value'].resample('YE').mean().plot()  # Resampling by year end ('YE')
plt.title('Average Technology Usage Over Time')
plt.xlabel('Year')
plt.ylabel('Usage Value')
plt.grid(True)
plt.show()

```

```python
# Compare usage patterns across different demographics (assuming 'group' or 'category')
plt.figure(figsize=(10, 6))
sns.boxplot(x='group', y='value', data=df)  # Replace 'group' with 'category' if more appropriate
plt.title('Technology Usage Across Different Groups')
plt.xlabel('Group')
plt.ylabel('Usage Value')
plt.grid(True)
plt.show()

```

```python
# Visualize the impact of technology on productivity
# Assuming 'value' corresponds to productivity score here:
plt.figure(figsize=(10, 6))
sns.scatterplot(x='value', y='value', hue='group', data=df)  # Replace columns as necessary
plt.title('Impact of Technology Usage on Productivity')
plt.xlabel('Technology Usage')
plt.ylabel('Productivity Score')
plt.grid(True)
plt.show()
```
