# Real Estate Price :

Analyze and visualize the distribution of real estate prices in different neighborhoods of a city. Use a dataset of real estate listings and create:

- A histogram to show the price distribution.

 -A box plot to compare price distributions across different neighborhoods.
 
- Scatter plots to show the relationship between property size and price.

__Hint:__ Use `plt.hist()` for histograms, `plt.boxplot()` for box plots, and `plt.scatter()` for scatter plots.

__Expectation:__ Visualizations showing the distribution of real estate prices, comparisons across neighborhoods, and the relationship between property size and price.


```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("AmesHousing.csv")

# Histogram creation
plt.figure(figsize = (10,6))
plt.hist(data['SalePrice'], edgecolor = "k")
plt.xlabel("Saleprice")
plt.ylabel("Frequency")
plt.title("Distribution of Sale Prices")
plt.show

# Box plot compare SalePrice distributions across Neighborhoods

plt.figure(figsize=(25,14))
sns.boxplot(x = 'Neighborhood', y = 'SalePrice', data = data,)
plt.xlabel('Neighborhood')
plt.ylabel('Saleprice')
plt.title('Sale Price Distribution Across Neighborhoods')
plt.grid()
plt.show()

# Scatter plot to show the relationship between property size and price

plt.scatter(data['Gr Liv Area'], data['SalePrice'], alpha = 0.5, edgecolor = 'w', linewidth = 0.5)
plt.xlabel("Area")
plt.ylabel("Saleprice")
plt.title("Relationship Between Property Size and Sale Price")
plt.grid()
plt.show()
```
